```python
auto_columns = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'model', 'origin', 'car name']
auto_data = pd.read_fwf('https://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data', header= None, names= auto_columns)
auto_data = auto_data.set_index("car name")
auto_data
```

**4. Some of the horsepower data is marked as missing with a ‘?’. You need to replace this ‘?’
with a pandas-appropriate missing value, then convert the column to numeric.**

I managed to fix the '?' in panadas with NaN using 'replace'.

```python
auto_data = auto_data.replace(to_replace='?', value=np.nan)
```

**5.The origin column values are 1, 2, and 3. Convert these values respectively to ‘USA’, ‘Asia’, and ‘Europe’**
selected column within the dataframe and replaced the 1,2, and 3 values.

```python
auto_data['origin'].replace(1, 'USA', inplace=True)
auto_data['origin'].replace(2, 'ASIA', inplace=True)
auto_data['origin'].replace(3, 'EUROPE', inplace=True)
```

```python
auto_data
```

**6.Create a bar chart that shows the distribution for cylinders.**

```python
auto_data.groupby("cylinders")
auto_data['cylinders'].plot(kind='barh', title="distribution of cylinders")
```

Added this extra bar graph to plot the data without the index.

```python
auto_data.groupby("cylinders").size().plot(kind="bar")
```

**7. Create a scatterplot that shows the relationship between horsepower and weight**

Used the scatterplot to show power to weight ration amongst cars.

```python
plt.scatter(auto_data['horsepower'], auto_data['weight'])
plt.suptitle('Horsepower and Weight')
plt.show()
```
